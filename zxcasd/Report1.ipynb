## 1.2 Разбиение сложных атрибутов

Сложными атрибутами можно считать атрибут автора, содержащий Фио отправителя и организацию отправителя. 

Разбиваю атрибут автора на ФИО самого автора и организацию автора

```python
temp = []
for elem in data["Автор"]:
    bracket_pos = elem.find('(')
    temp.append([elem[:bracket_pos-1], elem[bracket_pos+1:-1]])
```

```python
data[["Автор", "Организация Автора"]] = temp
```

## 1.3 Дополнение недостающими данными

Так как в текущем датасете авторами и адресатами являются конкретные лица, а не должности - составлю словарь замен, где ключ - ФИО, а значение - Должность(Организация которую представляет лицо)

```python
workers = dict(zip(data['Автор'], data['Организация Автора']))
```

```python
list(workers.items())[:5]
```

Также есть адресаты, которые не появляются в авторах, таковых можно считать за представителей ВСР. Так же им раздам временные роли, так как большая часть получателей сообщений - представители ВСР, а если таковых в модели будет слишком много, то модель сможет иметь большую точность всегда выдавая одного адресата

```python
temp_autors = data['Автор'].unique()
temp_recipients = data['Адресат'].unique()
```

```python
wsr_workers = [None if x in temp_autors else x for x in temp_recipients]
wsr_workers = list(filter(None, wsr_workers))
```

```python
wsr_workers = dict(zip(
    wsr_workers, 
    [f'Агентство развития профессиональных сообществ и рабочих кадров “Молодые профессионалы (Ворлдскиллс Россия) - Должность {i}' for i in range(len(wsr_workers))]))
```

```python
workers.update(wsr_workers)
```

```python
data[['Должность_адресата', 'Должность_автора']] = data[['Адресат', 'Автор']].replace(workers)
```

```python
data
```

Так как Имена Адресатов и Авторов уже заменены на их Должности(Организации), удаляю столбец с Организациями Авторов

## 1.4 Формирование словарей данных

### 1.4.1 Предобработка текста 

Перед формированием словарей данных стоит выполнить предобработку текста, так как текст находится в полном формате и имеет множество лишних символов, а также слова находятся не в начальной форме, что расширяет словарь при этом понижая количество полезной информации в нём

В качестве предобработки текста начала избавлюсь от всех символов кроме букв и пробелов, а также приведу текст к нижнему регистру, удалю все стоп слова (самые частовстречаемые и не имеющие какой-либо информации)

```python
data['Краткое\nсодержание'] = data['Краткое\nсодержание'].str.lower()\
                                                        .str.replace('[^a-zа-я ]', '', regex=True)
                                                        
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stopwords = stopwords.words('russian')
```

```python
data['Краткое\nсодержание'] = data['Краткое\nсодержание'].str.split()
```

```python
data['Краткое\nсодержание']
```

```python
import pymorphy2
```

```python
morph = pymorphy2.MorphAnalyzer(lang='ru')
```

```python
morph = pymorphy2.MorphAnalyzer(lang='ru')
word_replace = dict()
temp_col = []
for elem in data['Краткое\nсодержание']:
    temp_row = []
    for word in elem:
        if word not in word_replace.keys():
            word_replace[word] = morph.parse(word)[0].normal_form
        temp_row.append(word_replace[word])
    temp_col.append(' '.join(temp_row))
```

```python
data['Краткое\nсодержание'] = temp_col
```

### 1.4.2 Формирование словарей данных

Для формирования словарей данных преобразую текст в вид разряженной матрицы, где в названии столбца - слово, а в значении - количество его вхождений в строке, это позволит привести тексты в ормальный вид как для анализа ключевых слов, так и для последующего обучения моделей

Также при формировании атрибутов слов буду удалять те атрибуты чьё количество в датасете меньше 0.1% что позволит иметь ту же точность моделей при этом также уменьшая размерность датасета

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
CV = CountVectorizer(
    ngram_range = (1, 1),
    stop_words = stopwords.append('вх'),
    min_df = 0.001
).fit(data['Краткое\nсодержание'])
```

```python
import scipy.sparse
```

```python
vectorised_matrix = pd.DataFrame.sparse.from_spmatrix(
    CV.fit_transform(data['Краткое\nсодержание']),
    columns=CV.get_feature_names_out()
)
```

```python
vectorised_matrix
```

```python
data = pd.concat([data.reset_index(), vectorised_matrix], axis=1)
```

Ключевые слова успешно добавлены в датасет 

```python
data
```

Теперь удаляю сам столбец Краткого содержания, так как он уже преобразован в формат разряженной матрицы и не будет использоваться для классиикации

```python
data = data.drop(columns='Краткое\nсодержание')
```

Для анализа ключевых слов сгруппирую данные по должностям суммируя значения ключевых слов, а затем преобразуя всё в формат словаря, где ключ - должность, а значение - ключевые слова

```python
to_voc = vectorised_matrix.groupby(data['Должность_адресата'].reset_index(drop=True)).sum()
```

```python
to_voc = to_voc.T
```

```python
top_words = {}
for column in to_voc.columns:
    top_words[column] = to_voc[column].sort_values(ascending=False)
```

Сортирую должности по суммарной длине документов и вывожу сводку по главным ключевым словам, что позволит анализировать данные на предмет возможности обучения модели машинного обучения

```python
top_words = sorted(top_words.items(), key = lambda x: x[1].sum(), reverse = True)
```

```python
for addressee, words in top_words[:30]:
    print(f'Для должности: {addressee} Ключевые слова - \n  {list(words[words != 0].index[:10])}')
```

**По ключевым словам можно заметить, что для разных должностей соответствуют разные ключевые слова, а значит обучение моделей имеет смысл**

## 1.5 Преобразование списков переадресаций

За переадрессацию буду считать записи, где в комментарии есть полное ФИО адресата, для этого буду проверять есть ли в тексте комментария имя адресата, а затем заменю значения адресата в строчках

Выделяю все уникальные имена, для последующих проверок

```python
temp_h = data['Адресат'].append(data['Автор']).unique()
```

Функция для проверки есть ли какое-либо имя в тексте и возврашающая первое попавшееся имя

```python
def check_if_forwarding(text):
    text = str(text)
    for name in temp_h:
        if name in text:
            return name
    return None
```

Произвожу поиск и замену таких строк

```python
data_names = data_add.apply(check_if_forwarding)
```

```python
temp = []
for i, row in data[['index', 'Адресат']].iterrows():
    if data_names[row['index']] == None:
        temp.append(row['Адресат'])
    else: 
        temp.append(data_names[row['index']])
```

```python
data['Новый Адресат'] = temp
```

Проверяю в каких строках есть изменения, сравнивая Адресатов и Новых адресатов

```python
data[data['Новый Адресат'] != data['Адресат']]
```
