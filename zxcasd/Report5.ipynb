```python
app = Flask(__name__)

@app.route('/')
def index():
    return render_template('index.html')

@app.route('/', methods=['POST', 'GET'])
def upload_file():
    if request.method == 'POST':
        uploaded_file = request.files['file']
    data = pd.read_excel(uploaded_file)
    data['Адресат'] = bot_.model_predict(data['Краткое\nсодержание'])
    data['Адресат'] = data['Адресат'].apply(lambda x: sorted(x, key = lambda y: y[1], reverse=True)[0][0])
    template = render_template('test.html', data = data, addressee = bot_.model.classes_)
    return template

@app.route('/a', methods=['POST'])
def after_upload():
    data1 = pd.read_html(request.form['html'], header=0)[0]
    
    test = BeautifulSoup(open('Title.htm', 'rb'), 'html.parser')
    data1['Адресат'] = list(map(lambda x: x.find('option', selected='selected').text, test.find_all('select')))
    data1.drop(columns='true_prediction')
    towrite = io.BytesIO()
    data1.to_excel(towrite)  # write to BytesIO buffer
    test = send_file(towrite, download_name='REEEEEEEEEEEEEEEEEEEEEEEEEE.xlsx')
    towrite.seek(0)
    return test
    

@app.route('/test', methods=("POST", "GET"))
def html_table():

    return render_template('simple.html',  data = data, addressee = bot_.model.classes_)

if __name__ == '__main__':
    app.run()
```
